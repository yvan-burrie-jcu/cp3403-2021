We will now drop irrelevant columns, reduce outliers and export the preprocessed data.

```python
# The test date column is not needed.
data = data.drop(columns=['url','region_url','VIN','image_url','description','county','posting_date','model'])

# Only use prices over $500.
data = data.loc[data['price'] >= 500]

# Generate CSV file from dataframe.
data.to_csv(path_or_buf='preprocessedvehice2.csv', index=False)
```
